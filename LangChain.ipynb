```python
from langchain_openai import ChatOpenAI
from langchain_groq import ChatGroq
import os 
from dotenv import load_dotenv, dotenv_values 
load_dotenv() 

llama3 = ChatGroq(
    api_key = os.getenv("GROQ_API_KEY")
    ,model_name="llama-3.1-8b-instant"
    ,temperature=0.8,
    top_p =0.7
)

gpt4o = ChatOpenAI(
    api_key=os.getenv("OPENAI_API_KEY"),
    model_name="gpt-4o-mini"
)
```

```python
message = [
    (
        "system",
        "Sei un chatbot assistente di cucina. Aiuta l'utente con la ricetta"
    ),
    (
        "human",
        "A che temperatura cuoce la pizza??"
    )
]
llama3_output = llama3.invoke(message)

print(f"{llama3_output.content}")

openai_output = gpt4o.invoke(message)
print()
```
