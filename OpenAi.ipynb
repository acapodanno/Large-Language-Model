```python
from openai import OpenAI
client = OpenAI(api_key=OPENAPI_KEY)
```

```python
from inspect import trace
output = client.chat.completions.create(
    model="gpt-4.1-nano",
    messages=[
        {
            "role": "system",
            "content": "Ciao sei un italiano medio del boom economico e ti chiami Alessandro e rispondi sempre 'Che bella l'Italia'"
        },
        {
            "role": "user",
            "content": "Come ti chiami fra?? Eta?"
        }
    ],

)
print(output.choices[0].message.content)
```

### Function OpenAI

```python
import json

def search_badge_by_id(id):
  if(id == 1):
    print(f"The Badge {id} is of Alessandro")
  elif(id == 2):
    print(f"The Badge {id} is of Marco")

output = client.chat.completions.create(
    model="gpt-4.1-nano",
    messages=[
        {
            "role": "user",
            "content": "Ho trovato un badge in ufficio con numero 1"
        }
    ],
    functions=[
        {
            "name": "search_badge_by_id",
            "description": "Search badge by id",
            "parameters": {
                "type": "object",
                "properties": {
                    "id": {
                        "type": "integer",
                        "description": "The badge id"
                    }
                },
                "required": ["id"]
            }
        }
    ],
    function_call="auto"
)
print(f"Finish Reason: {output.choices[0].finish_reason}")
print(f"Message: {output.choices[0].message}")

if(output.choices[0].finish_reason == "function_call"):
    if output.choices[0].message.function_call.name == "search_badge_by_id":
        function_args = json.loads(output.choices[0].message.function_call.arguments)
        search_badge_by_id(**function_args)
```
